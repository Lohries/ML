1. **Bibliotecas**

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

2. **Visualizar**

```python
data_frame = pd.read_csv('/Users/lohries/Desktop/Python/Machine_Learning/Exercises/Csv/Tree.csv')
data_frame.head()
```

3. **Teste e Treino**

```python
X = data_frame.drop(['churn'], axis=1)
y = data_frame.drop(['id'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

4. **Modelo**

```python
classificador1 = DecisionTreeClassifier(random_state=42, max_depth=7) #1
classificador2 = DecisionTreeClassifier(random_state=42,min_samples_split=30, max_depth=7) #1+2
classificador3 = DecisionTreeClassifier(random_state=42, max_depth=7, min_samples_split=30, min_samples_leaf=12) #1+2+3
```

5. **Fitting**


```python
classificador1.fit(X_train, y_train)
classificador2.fit(X_train, y_train)
classificador3.fit(X_train, y_train)
```

6. **Predictions**

```python
y_train_prediction1 = classificador1.predict(X_train)
y_train_prediction2 = classificador2.predict(X_train)
y_train_prediction3 = classificador3.predict(X_train)
train_score_1 = accuracy_score(y_train, y_train_prediction1)
train_score_2 = accuracy_score(y_train, y_train_prediction2)
train_score_3 = accuracy_score(y_train, y_train_prediction3)

print(train_score_1)
print(train_score_2)
print(train_score_3)

y_test_predictions1 = classificador1.predict(X_test)
y_test_predictions2 = classificador2.predict(X_test)
y_test_predictions3 = classificador3.predict(X_test)
test_score_1 = accuracy_score(y_test, y_test_predictions1)
test_score_2 = accuracy_score(y_test, y_test_predictions2)
test_score_3 = accuracy_score(y_test, y_test_predictions3)

print(test_score_1)
print(test_score_2)
print(test_score_3)
```

7. **Visualizar a Arvore**

```python
tree.plot_tree(classificador3)

plt.show()
```
